**Cargar cada imagen colocando ceros en los canales verdes y azul y mostrarla**

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt

# Cargar la imagen
imagen = io.imread('enviorment.jpg')

# Crear una copia de la imagen para no modificar la original
imagen_roja = np.copy(imagen)

# Colocar ceros en los canales verde y azul
imagen_roja[:, :, 1] = 0  # Canal verde
imagen_roja[:, :, 2] = 0  # Canal azul

# Mostrar la imagen original y la imagen modificada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen)

plt.subplot(1, 2, 2)
plt.title('Solo Canal Rojo')
plt.imshow(imagen_roja)

plt.show()
```

**Aplanar cada imagen y grabarla en un archivo excel**

```python
# Cargar la imagen
imagen = io.imread('enviorment.jpg')

# Aplanar la imagen (convertirla en un arreglo unidimensional)
imagen_aplanada = imagen.flatten()

# Convertir el arreglo aplanado en un DataFrame de pandas
df = pd.DataFrame(imagen_aplanada, columns=['Valor de píxel'])

# Guardar el DataFrame en un archivo de Excel
df.to_excel('imagen_aplanada.xlsx', index=False)

print("La imagen aplanada se ha guardado en 'imagen_aplanada.xlsx'.")
```

**Cambiar el tamaño de cada imagen a un tamaño de 300x300 pixeles y mostrarla.**

```python
# Importamos la funcion rezise
from skimage.transform import resize

# Cargar la imagen
imagen = io.imread('enviorment.jpg')

# Cambiar el tamaño de la imagen a 300x300 píxeles
imagen_redimensionada = resize(imagen, (300, 300), preserve_range=True, anti_aliasing=True)

# Convertir la imagen redimensionada a tipo uint8 (opcional, para asegurar compatibilidad)
imagen_redimensionada = (imagen_redimensionada * 255).astype(np.)

# Mostrar la imagen original y la imagen redimensionada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen)

plt.subplot(1, 2, 2)
plt.title('Imagen Redimensionada (300x300)')
plt.imshow(imagen_redimensionada)

plt.show()
```
